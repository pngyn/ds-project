```python
#align with milestone notebook
import pandas as pd
import scipy as sc
import numpy as np

import statsmodels.formula.api as sm
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
%matplotlib inline  
plt.rcParams['figure.figsize'] = (10, 6) 

sample200 = pd.read_csv("../data/cleaned/200-language-sample.csv")
```

# Exploratory Analysis

```python
macroarea_count = sample200['macroarea'].value_counts()

#barchart of marco area counts
macroarea_count.plot(kind='bar', color='skyblue', ylabel='Count of Languages', xlabel='Macroarea', title='Count of Languages by Macroarea')
plt.xticks(rotation=45)
plt.show()
```

```python
#filter on countrycodes
country_count = sample200['countrycodes'].value_counts()
country_count
#barchart of country counts
country_count_top10 = country_count.head(10)
ax = country_count_top10.plot(kind='bar', color = 'salmon')
ax.set_ylabel('Count of Languages')
ax.set_xlabel('Country Codes')
ax.set_title('Top 10 Countries by Language Count')
xticks = plt.xticks(rotation=45)
ax.set_yticks(range(0, 24, 2))
plt.show()
```

```python
family = sample200['family'].value_counts()
multi_family = family[family > 2]
multi_family.plot(kind='barh', color='salmon', ylabel='Count of Languages', xlabel='Language Family', title='Count of Languages by Family (only families with more than 2 languages)')
plt.show()
```

```python
genus = sample200['genus'].value_counts()
multi_genus = genus[genus > 1]
multi_genus.plot(kind= 'barh', color='salmon', ylabel='Count of Genera', xlabel='Language Genus', title='Count of Languages by Genus (only families with more than 1 language')
plt.show()
```
